```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("training_jobs.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))

# Loop through the selected columns to create subplots
for i, column in enumerate(['model_size', 'dataset_size', 'batch_size', 'epochs', 'uses_gpu']):
    plt.subplot(2, 3, i+1)  # Adjust to 2x3 grid since there are 5 columns
    if df[column].dtype == 'bool' or df[column].dtype == 'object':  # If categorical, use countplot
        sns.countplot(x=df[column])
    else:  # Otherwise, plot a histogram
        sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'{column} Distribution')

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df[['model_size', 'dataset_size', 'batch_size', 'epochs',  'uses_gpu','true_runtime']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Features', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(15, 12))  

# Model Size vs True Runtime
plt.subplot(2, 3, 1)
plt.scatter(df['model_size'], df['true_runtime'], alpha=0.5)
plt.xlabel('Model Size')
plt.ylabel('True Runtime')
plt.title('Model Size vs True Runtime')

# Dataset Size vs True Runtime
plt.subplot(2, 3, 2)
plt.scatter(df['dataset_size'], df['true_runtime'], alpha=0.5)
plt.xlabel('Dataset Size')
plt.ylabel('True Runtime')
plt.title('Dataset Size vs True Runtime')

# Batch Size vs True Runtime
plt.subplot(2, 3, 3)
plt.scatter(df['batch_size'], df['true_runtime'], alpha=0.5)
plt.xlabel('Batch Size')
plt.ylabel('True Runtime')
plt.title('Batch Size vs True Runtime')

# Epochs vs True Runtime
plt.subplot(2, 3, 4)
plt.scatter(df['epochs'], df['true_runtime'], alpha=0.5)
plt.xlabel('Epochs')
plt.ylabel('True Runtime')
plt.title('Epochs vs True Runtime')

# Uses GPU vs True Runtime
plt.subplot(2, 3, 5)
plt.scatter(df['uses_gpu'], df['true_runtime'], alpha=0.5)
plt.xlabel('Uses GPU')
plt.ylabel('True Runtime')
plt.title('Uses GPU vs True Runtime')

plt.tight_layout()
plt.show()
```
